### Activity: Credit Card Churn

Objective: Process and visualize data to predict churn. In this scenario, a bank manager noticed an increasing number of clients terminating their credit card services with the bank. The goal is to predict which customers will leave so that the bank can proactively offer more appealing services to the client, thereby convincing them to reconsider.

Steps for the activity:
1. **Data Reading:** Using Pandas, read the database located in data/BankChunners. By the end of this step, the following questions can be answered:
  * What are the input attributes of the dataset?
  * What is the output attribute of the dataset?
  * How many clients are there?
  * Is there any correlation between the variables?
  
2. **Data Processing:** Check the status of the database, remove missing data, duplicates, and perform transformations on the attributes as necessary.

3. **Data Visualization:** Using Seaborn and Matplotlib for generate charts of the available data in various formats (e.g. barchart, pie chart, boxplot). Additionally, view data statistics.

#### Etapa 3 - Transformação dos dados não numéricos

```python
df.drop(['Avg_Open_To_Buy', 'Total_Trans_Ct', 'CLIENTNUM', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'],axis=1,inplace=True)
```

```python
df.columns
```

```python
df
```

```python
one_hot = pd.get_dummies(df[['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category']])
df = df.drop(['Gender', 'Education_Level', 'Marital_Status', 'Income_Category', 'Card_Category'],axis = 1)
df = df.join(one_hot)
df
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit_transform(df['Attrition_Flag'])
df['Attrition_Flag'] = le.fit_transform(df['Attrition_Flag'])
```
